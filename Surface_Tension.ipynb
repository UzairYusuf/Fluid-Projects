```python
import math

def capillary_height(surface_tension, contact_angle, density, gravity, distance):
  """Calculates the capillary height change between two parallel plates.

  Args:
    surface_tension: The surface tension of the liquid (N/m).
    contact_angle: The contact angle between the liquid and the plates (radians).
    density: The density of the liquid (kg/m).
    gravity: The acceleration due to gravity (m/s).
    distance: The distance between the plates (m).

  Returns:
    The capillary height change (m).
  """

  h = (2 * surface_tension * math.cos(contact_angle)) / (density * gravity * distance)
  return h

# for water at 20 degree C, W
surface_tension = 0.072  # N/m (surface tension of water at 20C)
contact_angle = 0  # radians (assuming a wetting contact angle)
density = 1000  # kg/m (density of water at 20C)
gravity = 9.81  # m/s (acceleration due to gravity)
distance = 0.5 * 10**-3  # m (distance between the plates)

height = capillary_height(surface_tension, contact_angle, density, gravity, distance)
print("Capillary height change:", height, "m")
```
